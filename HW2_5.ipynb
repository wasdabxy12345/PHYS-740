```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pylab as p
import matplotlib.animation as animation
from scipy.integrate import solve_ivp
import math

def shm1(t,u):
    omega0 = 3*math.pi
    omegaf = 2*math.pi
    tau = 2/omega0
    g1 = 1.061
    theta,omega = u
    return [omega,-(omega/tau) - (omega0**2)*math.sin(theta) + g1*(omega0**2)*math.cos(omegaf*t)]

def shm2(t,u):
    omega0 = 3*math.pi
    omegaf = 2*math.pi
    tau = 2/omega0
    g2 = 1.077
    theta,omega = u
    return [omega,-(omega/tau) - (omega0**2)*math.sin(theta) + g2*(omega0**2)*math.cos(omegaf*t)]

def shm3(t,u):
    omega0 = 3*math.pi
    omegaf = 2*math.pi
    tau = 2/omega0
    g3 = 1.081
    theta,omega = u
    return [omega,-(omega/tau) - (omega0**2)*math.sin(theta) + g3*(omega0**2)*math.cos(omegaf*t)]

def shm4(t,u):
    omega0 = 3*math.pi
    omegaf = 2*math.pi
    tau = 2/omega0
    g4 = 1.0826
    theta,omega = u
    return [omega,-(omega/tau) - (omega0**2)*math.sin(theta) + g4*(omega0**2)*math.cos(omegaf*t)]

def shm5(t,u):
    omega0 = 3*math.pi
    omegaf = 2*math.pi
    tau = 2/omega0
    g5 = 1.108
    theta,omega = u
    return [omega,-(omega/tau) - (omega0**2)*math.sin(theta) + g5*(omega0**2)*math.cos(omegaf*t)]

def shm6(tl,u):
    omega0 = 3*math.pi
    omegaf = 2*math.pi
    tau = 2/omega0
    g6 = 1.5
    theta,omega = u
    return [omega,-(omega/(2*tau)) - (omega0**2)*math.sin(theta) + g6*(omega0**2)*math.cos(omegaf*tl)]

mul = 100
t0, tf = 20.0, 50.0
nt = (tf - t0)*mul
t0l, tfl = 20.0, 3000.0
ntl = (tfl - t0l)*mul/5
t = np.linspace(t0, tf, int(nt)+1)
tl = np.linspace(t0l, tfl, int(ntl)+1)

u = [-(math.pi)/2,0.0]
sol1 = solve_ivp(shm1,[t0,tf],u,t_eval=t)
sol2 = solve_ivp(shm2,[t0,tf],u,t_eval=t)
sol3 = solve_ivp(shm3,[t0,tf],u,t_eval=t)
sol4 = solve_ivp(shm4,[t0,tf],u,t_eval=t)
sol5 = solve_ivp(shm5,[t0,tf],u,t_eval=t)
sol6 = solve_ivp(shm6,[t0l,tfl],u,t_eval=tl)

theta1 = sol1.y[0]
omega1 = sol1.y[1]
theta2 = sol2.y[0]
omega2 = sol2.y[1]
theta3 = sol3.y[0]
omega3 = sol3.y[1]
theta4 = sol4.y[0]
omega4 = sol4.y[1]
theta5 = sol5.y[0]
omega5 = sol5.y[1]
theta6 = sol6.y[0]
omega6 = sol6.y[1]

while np.amax(theta1)>np.pi:
    theta1=np.where(theta1>np.pi, theta1-2.0*np.pi, theta1)
while np.amin(theta1)<-np.pi:
    theta1=np.where(theta1<-np.pi, theta1+2.0*np.pi, theta1)

while np.amax(theta2)>np.pi:
    theta2=np.where(theta2>np.pi, theta2-2.0*np.pi, theta2)
while np.amin(theta2)<-np.pi:
    theta2=np.where(theta2<-np.pi, theta2+2.0*np.pi, theta2)

while np.amax(theta3)>np.pi:
    theta3=np.where(theta3>np.pi, theta3-2.0*np.pi, theta3)
while np.amin(theta1)<-np.pi:
    theta1=np.where(theta1<-np.pi, theta1+2.0*np.pi, theta3)

while np.amax(theta4)>np.pi:
    theta4=np.where(theta4>np.pi, theta4-2.0*np.pi, theta4)
while np.amin(theta4)<-np.pi:
    theta4=np.where(theta4<-np.pi, theta4+2.0*np.pi, theta4)

while np.amax(theta5)>np.pi:
    theta5=np.where(theta5>np.pi, theta5-2.0*np.pi, theta5)
while np.amin(theta1)<-np.pi:
    theta5=np.where(theta5<-np.pi, theta5+2.0*np.pi, theta5)

while np.amax(theta6)>np.pi:
    theta6=np.where(theta6>np.pi, theta6-2.0*np.pi, theta6)
while np.amin(theta6)<-np.pi:
    theta6=np.where(theta6<-np.pi, theta6+2.0*np.pi, theta6)

f1 = p.figure()
p.plot(t, theta1, label='Î“ = 1.061')
p.plot(t, theta2, label='Î“ = 1.077')
p.plot(t, theta3, label='Î“ = 1.081')
p.plot(t, theta4, label='Î“ = 1.0826')
p.plot(t, theta5, label='Î“ = 1.108')
#p.plot(tl, theta6, label='Î“ = 1.5')
p.grid()
p.legend(loc='best')
p.xlabel('time (sec)')
p.ylabel('angle (rad)')
f1.savefig('2_5(i).png')





f1 = p.figure()
plt.subplot(2, 3, 1)
p.plot(theta1, omega1, ',', label='Î“ = 1.061')
p.grid()
p.legend(loc='best')
p.xlabel('angle (rad)')
p.ylabel('angular velocity (rad/sec)')
plt.subplot(2, 3, 2)
p.plot(theta2, omega2, ',', label='Î“ = 1.077')
p.grid()
p.legend(loc='best')
p.xlabel('angle (rad)')
#p.ylabel('angular velocity')
plt.subplot(2, 3, 3)
p.plot(theta3, omega3, ',', label='Î“ = 1.081')
p.grid()
p.legend(loc='best')
p.xlabel('angle (rad)')
#p.ylabel('angular velocity')
plt.subplot(2, 3, 4)
p.plot(theta4, omega4, ',', label='Î“ = 1.0826')
p.grid()
p.legend(loc='best')
p.xlabel('angle (rad)')
p.ylabel('angular velocity (rad/sec)')
plt.subplot(2, 3, 5)
p.plot(theta5, omega5, ',', label='Î“ = 1.108')
p.grid()
p.legend(loc='best')
p.xlabel('angle (rad)')
#p.ylabel('angular velocity')
plt.subplot(2, 3, 6)
p.plot(theta6, omega6, ',', label='Î“ = 1.5')
p.grid()
p.legend(loc='best')
p.xlabel('angle (rad)')
#p.ylabel('angular velocity')
f1.savefig('2_5(ii).png')





td = t[0::int(nt/(tf-t0))]
tdl = tl[0::int(ntl/(tfl-t0l))]
sold1 = solve_ivp(shm1,[t0,tf],u,t_eval=td)
sold2 = solve_ivp(shm2,[t0,tf],u,t_eval=td)
sold3 = solve_ivp(shm3,[t0,tf],u,t_eval=td)
sold4 = solve_ivp(shm4,[t0,tf],u,t_eval=td)
sold5 = solve_ivp(shm5,[t0,tf],u,t_eval=td)
sold6 = solve_ivp(shm6,[t0l,tfl],u,t_eval=tdl)

thetad1 = sold1.y[0]
omegad1 = sold1.y[1]
thetad2 = sold2.y[0]
omegad2 = sold2.y[1]
thetad3 = sold3.y[0]
omegad3 = sold3.y[1]
thetad4 = sold4.y[0]
omegad4 = sold4.y[1]
thetad5 = sold5.y[0]
omegad5 = sold5.y[1]
thetad6 = sold6.y[0]
omegad6 = sold6.y[1]

while np.amax(thetad1)>np.pi:
    thetad1=np.where(thetad1>np.pi, thetad1-2.0*np.pi, thetad1)
while np.amin(thetad1)<-np.pi:
    thetad1=np.where(thetad1<-np.pi, thetad1+2.0*np.pi, thetad1)
    
while np.amax(thetad2)>np.pi:
    thetad2=np.where(thetad2>np.pi, thetad2-2.0*np.pi, thetad2)
while np.amin(thetad2)<-np.pi:
    thetad2=np.where(thetad2<-np.pi, thetad2+2.0*np.pi, thetad2)
    
while np.amax(thetad3)>np.pi:
    thetad3=np.where(thetad3>np.pi, thetad3-2.0*np.pi, thetad3)
while np.amin(thetad3)<-np.pi:
    thetad3=np.where(thetad3<-np.pi, thetad3+2.0*np.pi, thetad3)
    
while np.amax(thetad4)>np.pi:
    thetad4=np.where(thetad4>np.pi, thetad4-2.0*np.pi, thetad4)
while np.amin(thetad4)<-np.pi:
    thetad4=np.where(thetad4<-np.pi, thetad4+2.0*np.pi, thetad4)
    
while np.amax(thetad5)>np.pi:
    thetad5=np.where(thetad5>np.pi, thetad5-2.0*np.pi, thetad5)
while np.amin(thetad5)<-np.pi:
    thetad5=np.where(thetad5<-np.pi, thetad5+2.0*np.pi, thetad5)
    
while np.amax(thetad6)>np.pi:
    thetad6=np.where(thetad6>np.pi, thetad6-2.0*np.pi, thetad6)
while np.amin(thetad6)<-np.pi:
    thetad6=np.where(thetad6<-np.pi, thetad6+2.0*np.pi, thetad6)
    
f1 = p.figure()
plt.subplot(2, 3, 1)
p.plot(thetad1, omegad1, '.', label='Î“ = 1.061')
p.grid()
p.legend(loc='best')
p.xlabel('angle (rad)')
p.ylabel('angular velocity (rad/sec)')
plt.subplot(2, 3, 2)
p.plot(thetad2, omegad2, '.', label='Î“ = 1.077')
p.grid()
p.legend(loc='best')
p.xlabel('angle (rad)')
#p.ylabel('angular velocity')
plt.subplot(2, 3, 3)
p.plot(thetad3, omegad3, '.', label='Î“ = 1.081')
p.grid()
p.legend(loc='best')
p.xlabel('angle (rad)')
#p.ylabel('angular velocity')
plt.subplot(2, 3, 4)
p.plot(thetad4, omegad4, '.', label='Î“ = 1.0826')
p.grid()
p.legend(loc='best')
p.xlabel('angle (rad)')
p.ylabel('angular velocity (rad/sec)')
plt.subplot(2, 3, 5)
p.plot(thetad5, omegad5, '.', label='Î“ = 1.108')
p.grid()
p.legend(loc='best')
p.xlabel('angle (rad)')
#p.ylabel('angular velocity')
plt.subplot(2, 3, 6)
p.plot(thetad6, omegad6, '.', label='Î“ = 1.5')
p.grid()
p.legend(loc='best')
p.xlabel('angle (rad)')
#p.ylabel('angular velocity')
f1.savefig('2_5(iii).png')
```
