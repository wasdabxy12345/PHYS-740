```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
#matplotlib.use('tkagg')
import matplotlib.pyplot as plt
from numpy import *
import pylab as p
from scipy import integrate


y82 = [8,2] # [rabbit, fox]
y92 = [9,2] # [rabbit, fox]
y72 = [7,2] # [rabbit, fox]
y83 = [8,3] # [rabbit, fox]
y81 = [8,1] # [rabbit, fox]

t0, tf = 0.0, 25
t = np.linspace(t0, tf, 26)

alpha = 1.0
beta = 0.5
gamma = 0.5
delta = 4.0

params = [alpha, beta, gamma, delta]
 
def sim(t, variables, params):
    
    #rabbit population level
    x = variables[0]
    
    # fox population level
    y = variables[1]
    
    
    alpha = params[0]
    beta = params[1]
    gamma = params[2]
    delta = params[3]
    
    dxdt = alpha * x - beta * x * y
    dydt = gamma * x * y - delta * y
    
    return([dxdt, dydt])


y_82 = solve_ivp(sim, [t0, tf], y82, t_eval=t, args=(params,))
y_92 = solve_ivp(sim, [t0, tf], y92, t_eval=t, args=(params,))
y_72 = solve_ivp(sim, [t0, tf], y72, t_eval=t, args=(params,))
y_83 = solve_ivp(sim, [t0, tf], y83, t_eval=t, args=(params,))
y_81 = solve_ivp(sim, [t0, tf], y81, t_eval=t, args=(params,))

r82, f82 = y_82.y
r92, f92 = y_92.y
r72, f72 = y_72.y
r83, f83 = y_83.y
r81, f81 = y_81.y

f1 = p.figure()
p.plot(t, f82,'g-', label='N1(0) = 8; N2(0) = 2')
p.plot(t, r82,'g-')
p.plot(t, f92,'r-', label='N1(0) = 9; N2(0) = 2')
p.plot(t, r92,'r-')
p.plot(t, f72,'b-', label='N1(0) = 7; N2(0) = 2')
p.plot(t, r72,'b-')
p.plot(t, f83,'y-', label='N1(0) = 8; N2(0) = 3')
p.plot(t, r83,'y-')
p.plot(t, f81,'c-', label='N1(0) = 8; N2(0) = 1')
p.plot(t, r81,'c-')
p.grid()
p.legend(loc='best')
p.xlabel('time (weeks)')
p.ylabel('population (thousands)')
#p.title('Evolution of fox and rabbit populations')
f1.savefig('2_1(b).png')
```
