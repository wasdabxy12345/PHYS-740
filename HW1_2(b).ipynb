```python
import numpy as np
import math
import matplotlib.pyplot as plt

r = 1000
n = []
H = []
T = []
S = []
e_t = []
e_s = []

for i in range(2,r):
    n.append(i)

for i in range(r-2):
    t = 0
    p = np.linspace(0,(2*math.pi),n[i])   
    for j in range(n[i]-1):
        h = p[1]
        t += ((((math.sin(p[j]))**2) + ((math.sin(p[j+1]))**2)) * (h / 2))
    H.append(h)    
    T.append(t)

for i in range(r-2):
    e_t.append(abs((math.pi) - T[i]))

for i in range(r-2):
    s = 0
    p = np.linspace(0,(2*math.pi),n[i]) 
    for j in range(n[i]-1):
        h = p[1]
        s += ((((math.sin(p[j]))**2)+((math.sin(p[j+1]))**2)+(4*((math.sin((p[j]+p[j+1])/2))**2)))*(h/6))
    S.append(s)

for i in range(r-2):
    e_s.append(abs((math.pi) - S[i]))

fig, ax = plt.subplots(figsize = (35, 23))
plt.rcParams['font.size'] = 30
ax.scatter(H, e_t, label='Error when using the trapezoidal rule')
ax.scatter(H, e_s, label="Error when using the Simpson's rule")
#ax.plot(H, e_t)
#ax.plot(H, e_s)
ax.legend()
ax.set_xscale("log")
ax.set_yscale("log")
plt.xlabel("step size")
plt.ylabel("error")
plt.grid()
plt.savefig('2(b)')
plt.show()
```
