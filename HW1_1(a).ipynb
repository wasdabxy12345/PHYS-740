```python
import numpy as np
import math
import matplotlib.pyplot as plt

s = 300
p = np.linspace(0,10,s)

h = []
der2 = []
der4 = []
der6 = []
der8 = []
err2 = []
err4 = []
err6 = []
err8 = []

for i in range(s):
    h.append(1/(10**p[i]))
    
for i in range(s):
    der2.append((math.exp(h[i])-math.exp(-h[i]))/(2*h[i]))
for i in range(s):
    err2.append(abs(1-der2[i]))
    
for i in range(s):
    der4.append(((math.exp(-2*h[i]))/12-(math.exp(-h[i]))*2/3+(math.exp(h[i]))*2/3-(math.exp(2*h[i]))/12)/(h[i]))
for i in range(s):
    err4.append(abs(1-der4[i]))
    
for i in range(s):
    der6.append(((-math.exp(-3*h[i]))/60+(math.exp(-2*h[i]))*3/20-(math.exp(-h[i]))*3/4+(math.exp(h[i]))*3/4-(math.exp(2*h[i]))*3/20+(math.exp(3*h[i]))/60)/(h[i]))
for i in range(s):
    err6.append(abs(1-der6[i]))

for i in range(s):
    der8.append(((math.exp(-4*h[i]))/280-(math.exp(-3*h[i]))*4/105+(math.exp(-2*h[i]))/5-(math.exp(-h[i]))*4/5+(math.exp(h[i]))*4/5-(math.exp(2*h[i]))/5+(math.exp(3*h[i]))*4/105-(math.exp(4*h[i]))/280)/(h[i]))
for i in range(s):
    err8.append(abs(1-der8[i]))

s2 = s - 151
s4 = s - 212
s6 = s - 243
s8 = s - 257
dx_2 = h[s2] - h[0]
dx_4 = h[s4] - h[0]
dx_6 = h[s6] - h[0]
dx_8 = h[s8] - h[0]
dy_2 = err2[s2]-err2[0]
dy_4 = err4[s4]-err4[0]
dy_6 = err6[s6]-err6[0]
dy_8 = err8[s8]-err8[0]

sl_2 = dy_2/dx_2
sl_4 = dy_4/dx_4
sl_6 = dy_6/dx_6
sl_8 = dy_8/dx_8
print("slope of 2nd order = ",sl_2)
print("slope of 4th order = ",sl_4)
print("slope of 6th order = ",sl_6)
print("slope of 8th order = ",sl_8)

point1_2 = [h[0], err2[0]]
point2_2 = [h[s2], err2[s2]]

x_value_2 = [point1_2[0], point2_2[0]]
y_value_2 = [point1_2[1], point2_2[1]]

point1_4 = [h[0], err4[0]]
point2_4 = [h[s4], err4[s4]]

x_value_4 = [point1_4[0], point2_4[0]]
y_value_4 = [point1_4[1], point2_4[1]]

point1_6 = [h[0], err6[0]]
point2_6 = [h[s6], err6[s6]]

x_value_6 = [point1_6[0], point2_6[0]]
y_value_6 = [point1_6[1], point2_6[1]]

point1_8 = [h[0], err8[0]]
point2_8 = [h[s8], err8[s8]]

x_value_8 = [point1_8[0], point2_8[0]]
y_value_8 = [point1_8[1], point2_8[1]]

fig, ax = plt.subplots(figsize = (35, 23))
plt.rcParams['font.size'] = 30
ax.scatter(h, err2, label='2nd order')
ax.scatter(h, err4,label='4th order')
ax.scatter(h, err6,label='6th order')
ax.scatter(h, err8,label='8th order')
ax.plot(x_value_2, y_value_2, label='slope of 2nd order = 0.17520301443323494')
ax.plot(x_value_4, y_value_4, label='slope of 4nd order = 0.03758465148458164')
ax.plot(x_value_6, y_value_6, label='slope of 6nd order = 0.008781782592269412')
ax.plot(x_value_8, y_value_8, label='slope of 8nd order = 0.002162424355689539')
ax.legend()
#txt="Fig. 1(a): Plots of error vs step size for 2nd, 4th, 6th and 8th order central difference for the derivative of f(x) = exp(x) at x = 0"
#plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center')
ax.set_xscale("log")
ax.set_yscale("log")
plt.xlabel("Step Size")
plt.ylabel("Error")
plt.grid()
plt.savefig('1(a)')
plt.show()
```
