```python
import numpy as np
import math
import matplotlib.pyplot as plt

r = 100
n = []
H = []
T = []
S = []
e_t = []
e_s = []

for i in range(2,r):
    n.append(i)

for i in range(r-2):
    t = 0
    p = np.linspace(0,1,n[i])   
    for j in range(n[i]-1):
        h = p[1]
        t += (math.exp(p[j]) + math.exp(p[j+1])) * h / 2
    H.append(h)    
    T.append(t)

for i in range(r-2):
    e_t.append(abs((np.exp(1)-1) - T[i]))

for i in range(r-2):
    s = 0
    p = np.linspace(0,1,n[i])   
    for j in range(n[i]-1):
        h = p[1]
        s += (math.exp(p[j])+math.exp(p[j+1])+4*math.exp((p[j]+p[j+1])/2))*h/6
    S.append(s)

for i in range(r-2):
    e_s.append(abs((np.exp(1)-1) - S[i]))

dx = H[r-3] - H[0]
dy_t = e_t[r-3]-e_t[0]
dy_s = e_s[r-3]-e_s[0]

sl_t = dy_t/dx
sl_s = dy_s/dx
print("slope of integration using trapezoidal rule = ",sl_t)
print("slope of integration using Simpson's method = ",sl_s)

point1 = [H[0], e_t[0]]
point2 = [H[r-3], e_t[r-3]]

x_values = [point1[0], point2[0]]
y_values = [point1[1], point2[1]]

point1s = [H[0], e_s[0]]
point2s = [H[r-3], e_s[r-3]]

x_valuess = [point1s[0], point2s[0]]
y_valuess = [point1s[1], point2s[1]]

fig, ax = plt.subplots(figsize = (35, 23))
plt.rcParams['font.size'] = 30
ax.scatter(H, e_t, label='Error when using the trapezoidal rule')
ax.scatter(H, e_s, label="Error when using the Simpson's rule")
ax.plot(x_values, y_values, label='Slope of error(trapezoidal rule) = 0.14229617818596496')
ax.plot(x_valuess, y_valuess, label = "Slope of error(Simpson's rule) = 0.0005852958173791449")
ax.legend()
#txt="Fig. 1(a): Plots of error vs step size for 2nd, 4th, 6th and 8th order central difference for the derivative of f(x) = exp(x) at x = 0"
#plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center')
ax.set_xscale("log")
ax.set_yscale("log")
plt.xlabel("step size")
plt.ylabel("error")
plt.grid()
plt.savefig('2(a)')
plt.show()
```
